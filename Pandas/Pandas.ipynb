### Converting multiple arrays into one df

```python
# Arrays creation
ar1 = np.random.rand(3,4)
ar2 = np.random.rand(3,4)
ar3 = np.random.rand(3,4)
```

```python
# DFs creation
df1 = pd.DataFrame(ar1, columns = ['a', 'b', 'c', 'd'], index=[0,1,2])
df2 = pd.DataFrame(ar2, columns = ['a', 'b', 'c', 'd'], index=[0,1,2])
df3 = pd.DataFrame(ar3, columns = ['a', 'b', 'c', 'd'], index=[0,1,2])
```

```python
# Concatinating dfs
df4 = pd.concat([df1, df2, df3]) #row wise concatination
df4
```

```python
# Concatinating dfs
df5 = pd.concat([df1, df2, df3], axis=1) #column wise concatination
df5
```

We have a duplicate name indexes, but python has its built-in, default indexes. 
We can use iloc to select indexes based on default number.
Or we can use loc with name indexes, but in this case it will return multipe results. 

```python
df4.iloc[0]
```

```python
df4.loc[0]
```

## DF Manipulation

### Column creation based on existing columns

```python
df = pd.DataFrame(np.random.rand(10,4), columns=['a','b', 'c', 'd'], index=pd.date_range('2021-07-18', periods=10))
df
```

```python
# Basic operation columns creation
df['new'] = df['c'] - df['d']
```

```python
# Column creation using apply and lambda
df['new_col_squared'] = df['d'].apply(lambda x: x **2)
```

```python
df
```

```python
# log column of column d by using apply and labmda
df['log'] = df['d'].apply(lambda x: np.log(x))
```

```python
df
```

```python
# Column creation based on outside function

def test(x):
    if x < 0:
        return x
    else:
        return np.log(x)
```

```python
df['new_col_log_outside_func'] = df['d'].apply(test)
```

```python
df
```
