### Clustering

* KMeans
* Hierarchical
* DBSCAN

Why use?
    > When supervised ML algorithms we don't achieve satisfactory results, there is a high probability that simple supervised ML algorithms can't detect the pattern in the data. Clustering helps to group the data. Upon grouped data the supervised ML algorithms can be applied.

Application:
* customer segmentation,
* data analysis,
* outlier detection (anomaly detection),
* search engine, 
* speech/image segmentation.

### KMeans (divisive approach)

Finding value of k by using WCSS (within cluster summation of square or inertia) - calculating (euclidean)distance of centroid for all the data points. 
To evaluate the best value of k, we are creating elbow plot. In x ax we have # of k, on y ax we have wcss value. On this graph we observe, where no major changes will happen. 


On the plot we can see blue vertical line, which marks the best point of k/wcss.


Drawback of KMeans:
    > allocation of the centroid - randomly generated points will always take the nearest possible points, meaning that centroids cannot have possibility to create their own cluster. But there aresolutions for that -> KMean ++


### KMean ++

What's the difference?
    > Placing centroids in random way, but in some distances.
    
! Using sklearn we by default implement KMean++
