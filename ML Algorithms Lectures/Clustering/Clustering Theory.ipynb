### KMeans (divisive approach)

Finding value of k by using WCSS (within cluster summation of square or inertia) - calculating (euclidean)distance of centroid for all the data points. 
To evaluate the best value of k, we are creating elbow plot. In x ax we have # of k, on y ax we have wcss value. On this graph we observe, where no major changes will happen. 


On the plot we can see blue vertical line, which marks the best point of k/wcss.


Drawback of KMeans:
    > allocation of the centroid - randomly generated points will always take the nearest possible points, meaning that centroids cannot have possibility to create their own cluster. But there aresolutions for that -> KMean ++

### DBSCAN - Density Based Spatial Clustering of Application with Noise

Hyperparameters (value of is determined by the experiments):
* Epsilon - distance
* Min. Points - min number of points in particular cluster
* Core Points - which is already a part of cluster and it is used to form another cluster (points which satisfied min points criteria)
* Border Points - points which are part of some cluster, but will not form other clusters
* Noise - points which are not a part of any of the clusters (outliers)

We try to define Eps, MinPoints, 

Algorithm is starting from any random point which is a part of the dataset. From this point it is trying to draw a circle of the radius equal to the value of the epsilon. It is calculating how many points are inside this circle and check if the the min points criteria is satisfied. If the min points criteria is satisifed, the cluster can be formed. If not, it knows that this data point (from which we draw a circle) is not eligble to make a cluster. Algorithm checks the same procedure with other point. 

If the number of points is bigger than the min points criteria, the cluster will be formed. 

When the algorithm forms the first cluster, it tries to form another cluster from the points, which are part of the cluster. Point which is already a part of cluster and it is used to form another cluster is called **core point**. 



**DBSCAN can detect outliers**. The points that are not part of any cluster, are considered as an outliers. 

Drawbacks:
    > Extreme values of EPS and MinPoints will make DBSCAN to fail.

### How to evaluate clustering?

By the value of the interclass similarity and intraclass similarity (low value of mistakes, low value of ungroupped datapoints).

https://www.analyticsvidhya.com/blog/2020/10/quick-guide-to-evaluation-metrics-for-supervised-and-unsupervised-machine-learning/
https://towardsdatascience.com/performance-metrics-in-machine-learning-part-3-clustering-d69550662dc6

**Evaluation with a confusion matrix**

ss -> true positives - data points that are both placed in the same cluster, as in the provided labels
dd -> true negatives - data points that are neither a part of the cluster in the algo, and neither a part of the cluster in the provided labels
sd -> false positives predictions - in the ground truth data point is a NOT part of the cluster, but the algo says it is. 
ds -> false negatives predictions - in the ground truth data point is a part of the cluster, but the algo says it is NOT. 

* Rand Index - SS+DD/SS+DD+sd+ds
* Jaccard Coefficient - SS/ss+sd+ds
* Entropy - -SUM Pi log(Pi) - how random the distrbution is
* Purity - SS/Total number of the data points - Percentage of the data points classified correctly
* Silhoutte Coefficient - (b(x) - a(X))/max (a,b)

a(x) - avg distance of x from other data points in the same cluster
b(x) - avg distance of x from all the points in the next nearest cluster


### PCA - Principal Component Analysis - Dimensionality Reduction

Data preprocessing technique. It is not used to train. It is used to reduce (not remove). By using PCA we do not eliminate the columns, but convert the data in such a way that it will reatain it's original meaning.

Many features - why reduce the number of dimensions?
    * Time
    * Difficult to generalize 
    * Not all the features will contribute in making predictions
    * Hard to find out shape of the data
    
PCA says:
    * Take the data which have mean equal to 0 and std equal to 1. Meaning, the dataset needs to be performed into Standard Normal Distribution (z = (x-mean)/std). 
        > ! PCA works only on Standard Normal Distribution data. Not Normal Distributed data.
    * Draw a single straight line, which will the best explain the data (like best fitted line in Linear Regression - to make the residuals minimum). This line is called **Principal Component Line**. 
    * This PC1 line rotate to the place of x axis. The points represented on the new line PC1, can create new column. 
    * Next draw PC2 line, which is a horizontal line in the place of y axis etc.
    
How many PC lines should we draw?
    > Explained Variance Ratio (EVR) - how many relathionships single PC line is capable to capture; distance of PC1 Points/ Distance of PC1 + Distance of PCn. The value of EVR tells us what is the percentage of retain meaning of the data. 
    

    
    > We evaluate the number of PC lines we need, with screen plot.
