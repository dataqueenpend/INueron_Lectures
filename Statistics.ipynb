### Imports

```python
import numpy as np
import statsmodels.api as sm
import pylab as py
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## Random data generation

```python
# Standard normal distribution - generating random data
data_points = np.random.normal(0,1,100)
```

```python
# Log normal distirbution - generating random data
gfg = np.random.lognormal(0, 1, 100)
```

```python
# Pareto distribution - generating random data
pdr = np.random.pareto(1, 100)
```

```python
# Non-normal(exponential) distribution - generating random data
exp = np.random.exponential(size = 100)

# Transforming data & saving lambda value
d, dl = stats.boxcox(exp)
```

## QQPLOT

```python
# QQ-plot - standard normal distribution
sm.qqplot(data_points, line='45')
py.show()
```

```python
#QQ-plot - Log normal distrbution
sm.qqplot(gfg, line='45')
py.show()
```

```python
# QQ-plot - Pareto distrbution
sm.qqplot(pdr, line='45')
py.show()
```

```python
# QQ-plot - Pareto distrbution (from np exponential)
sm.qqplot(exp, line='45')
py.show()
```

```python
# QQ- plot - Exponential distrbution (pareto) after box-cox transformation
sm.qqplot(d, line='45')
py.show()
```

```python
# Creating axes
fig, ax = plt.subplots(1,2)

# Drawing exponential and fitted (normal) data (after transfromation)
sns.distplot(exp, hist=False, kde=True,
           ax=ax[0], label ='Non-normal',
            kde_kws = {'shade': True, 'linewidth': 2}, color='blue')
sns.distplot(d, hist=False, kde=True,
            ax=ax[1], label='Normal',
            kde_kws = {'shade': True, 'linewidth': 2}, color='yellow')

# Rescaling subplots
fig.set_figheight(5)
fig.set_figwidth(10)

# Adding legends
plt.legend(loc='upper right')
```
